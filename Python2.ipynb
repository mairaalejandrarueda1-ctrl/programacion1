**DESDE LISTA**

```python
# lista de frutas
frutas = ["manzana", "pera", "uva", "manzana", "mango"]

print("lista original:", frutas)

# Lista a tupla
frutas_tupla = tuple(frutas)
print("Convertida a tupla:", frutas_tupla)

# Lista a diccionario
frutas_diccionario = dict(enumerate(frutas))
print("Convertida a diccionario", frutas_diccionario)

# Lista a conjunto
frutas_conjunto = set(frutas)
print("Convertida a conjunto:", frutas_conjunto)
```

**DESDE TUPLA**

```python
# Tubla de utiles escolares
utiles = ("cuaderno", "lapiz", "lapicero" "lapiz", "borrador", "carpeta")

print("Tupla original:", utiles)

# Tupla a lista
utiles_lista = list(utiles)
print("Convertida a diccionario:", utiles_lista)

# Tupla a diccionario
utiles_diccionario = dict(enumerate(utiles))
print("Convertida a diccionario:", utiles_diccionario)

# Tupla a conjunto
utiles_conjunto = set(utiles)
print("Convertida a conjunto:", utiles_conjunto)
```

**DESDE DICCIONARIO**

```python
# Diccionario con estudiante y carrera
estudiante = {"nombre": "Laura", "carrera": "Ingenieria", "semestre": "3"}
print("Diccionario original:", estudiante)

# Diccionario a lista
estudiante_lista = list(estudiante.items())
print("Convertido a lista:", estudiante_lista)

# Diccionario a tupla
estudiante_tupla = tuple(estudiante.items())
print("Convertido a tupla:", estudiante_tupla)

# Diccionario a conjunto
estudiante_conjunto = set(estudiante.items())
print("Convertido a conjunto:", estudiante_conjunto)
```

**DESDE CONJUNTO**

```python
# Conjunto de colores
colores = {"azul", "verde", "rojo", "amarillo"}

print("Conjunto original:", colores)

# Conjunto a lista
colores_lista = list(colores)
print("Convertido a lista:", colores_lista)

# Conjunto a tupla
colores_tupla = tuple(colores)
print("Convertido a tupla:", colores_tupla)

# Conjunto a diccionario
colores_diccionario = dict(enumerate(colores))
print("Convertido a diccionario:", colores_diccionario)
```
