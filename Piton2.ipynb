```python
lista = [1, 2, 3, 4]

print("Lista original:", lista)
tupla_desde_lista = tuple(lista)
print("Lista a tupla:", tupla_desde_lista)
conjunto_desde_lista = set(lista)
print("Lista a conjunto:", conjunto_desde_lista)
diccionario_desde_lista = dict(enumerate(lista))
print("Lista a diccionario:", diccionario_desde_lista)
```

```python
tupla = (5, 6, 7, 8)

print("\nTupla original:", tupla)
lista_desde_tupla = list(tupla)
print("Tupla a lista:", lista_desde_tupla)
conjunto_desde_tupla = set(tupla)
print("Tupla a conjunto:", conjunto_desde_tupla)
diccionario_desde_tupla = dict(enumerate(tupla))
print("Tupla a diccionario:", diccionario_desde_tupla)
```

```python
diccionario = {"a": 1, "b": 2, "c": 3}

print("\nDiccionario original:", diccionario)
lista_desde_diccionario = list(diccionario)
print("Diccionario a lista:", lista_desde_diccionario)
tupla_desde_diccionario = tuple(diccionario)
print("Diccionario a tupla:", tupla_desde_diccionario)
conjunto_desde_diccionario = set(diccionario)
print("Diccionario a conjunto:", conjunto_desde_diccionario)
```

```python
conjunto = {9, 10, 11}

print("\nConjunto original:", conjunto)

lista_desde_conjunto = list(conjunto)
print("Conjunto a lista:", lista_desde_conjunto)

tupla_desde_conjunto = tuple(conjunto)
print("Conjunto a tupla:", tupla_desde_conjunto)

diccionario_desde_conjunto = dict(enumerate(conjunto))
print("Conjunto a diccionario:", diccionario_desde_conjunto)
```
